```python

# Importazione delle librerie
##https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris
from sklearn.datasets import load_iris

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, svm
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split

def load_iris_data():
    # Caricamento del dataset Iris
    iris = datasets.load_iris()
    X = iris.data
    y = iris.target
    class_names = iris.target_names
    return X, y, class_names

# Caricamento dei dati
X, y, class_names = load_iris_data()

# Divisione del dataset in set di addestramento
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Creazione del classificatore SVM con parametro di regolarizzazione di C=0.01
classifica = svm.SVC(kernel="linear", C=0.01).fit(X_train, y_train)

# Impostazione delle opzioni di visualizzazione della matrice di confusione
#Matrice di Confusione Non Normalizzata:
#I valori nella matrice rappresentano il numero totale di campioni che sono stati classificati in modo specifico.

#Matrice di Confusione Normalizzata:
#I valori nella matrice rappresentano la percentuale di campioni rispetto al totale.
np.set_printoptions(precision=2)
titoli = [
    ("Matrice di confusione senza normalizzazione", None),
    ("Matrice di confusione normalizzata", "true"),
]

# Iterazione tra le opzioni di visualizzazione della matrice di confusione
#Cicla 2 volte (lunghezza del vettore)
for i, normalizzata in titoli:
    # Creazione e visualizzazione della matrice di confusione
    disp = ConfusionMatrixDisplay.from_estimator(classifica,X_test,y_test,
        display_labels=class_names, cmap=plt.cm.Blues, normalizzata=normalizzata)
    disp.ax_.set_title(i)

    # Stampa della matrice di confusione
    print(i)
    print(disp.confusion_matrix)

# Visualizzazione delle matrici di confusione
plt.show()
```
